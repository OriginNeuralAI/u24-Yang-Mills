# 05 — E_8 Subgroup Traces

Verify SU(3) is the unique compact simple Lie group with Tr(J) = 24.

```python
import matplotlib.pyplot as plt

groups = {
    'SU(2)': (2, 3), 'SU(3)': (3, 8), 'G2': (4, 14), 'SU(4)': (4, 15),
    'SU(5)': (5, 24), 'SO(5)': (3, 10), 'SO(10)': (8, 45),
    'F4': (9, 52), 'E6': (12, 78), 'E7': (18, 133), 'E8': (30, 248)
}

names, traces = [], []
for name, (c2, dim) in groups.items():
    tr = c2 * dim
    names.append(name); traces.append(tr)
    marker = ' <-- OMEGA' if tr == 24 else ''
    print(f'{name}: C2={c2}, dim={dim}, Tr={tr}{marker}')

colors = ['#D4AF37' if t == 24 else '#2A6482' for t in traces]
plt.barh(names, traces, color=colors)
plt.axvline(x=24, color='#D4AF37', linestyle='--', linewidth=2, label='Omega = 24')
plt.xscale('log'); plt.xlabel('Tr(J_G)'); plt.legend()
plt.title('Only SU(3) has Tr(J) = Omega = 24'); plt.tight_layout(); plt.show()
```
