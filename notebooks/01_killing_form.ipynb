# 01 — The Killing Form Identity

Verify that Tr(J_SU(3)) = 24 = Omega from the SU(3) structure constants.

```python
import numpy as np

# SU(3) structure constants (Gell-Mann basis)
f = np.zeros((8, 8, 8))
for a, b, c, v in [
    (1,2,3, 1.0), (1,4,7, 0.5), (1,5,6, -0.5),
    (2,4,6, 0.5), (2,5,7, 0.5), (3,4,5, 0.5),
    (3,6,7, -0.5), (4,5,8, np.sqrt(3)/2), (6,7,8, np.sqrt(3)/2),
]:
    a, b, c = a-1, b-1, c-1
    for i,j,k,s in [(a,b,c,v),(b,c,a,v),(c,a,b,v),(b,a,c,-v),(a,c,b,-v),(c,b,a,-v)]:
        f[i,j,k] = s

J = np.einsum('acd,bcd->ab', f, f)
print(f'J_SU(3) = {J[0,0]:.1f} * I_8')
print(f'Eigenvalues: {np.linalg.eigvalsh(J)}')
print(f'Tr(J) = {np.trace(J):.1f}')
print(f'Omega = 24')
print(f'MATCH: {np.isclose(np.trace(J), 24)}')
```

```python
# Uniqueness: check all SU(N) for N=2..20
print('SU(N) Killing form traces:')
for N in range(2, 21):
    tr = N * (N**2 - 1)
    marker = ' <-- Omega!' if tr == 24 else ''
    print(f'  SU({N}): Tr = {tr}{marker}')
```
