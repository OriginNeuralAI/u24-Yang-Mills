# 04 — OS Axiom Verification

Check all 5 Osterwalder-Schrader axioms from engine data.

```python
import json
with open('../data/os-axioms/os_axiom_verification.json') as f:
    data = json.load(f)

print('Free energy convergence:')
for key, val in data['free_energy_convergence'].items():
    print(f"  {key}: change = {val['relative_change']:.4f}, converged = {val['converged']}")

print(f"
Correlator decay: Delta_eff = {data['correlator_decay']['effective_mass_gap']:.3f}")
print(f"Decay confirmed: {data['correlator_decay']['decay_confirmed']}")

print(f"
L-independence: delta_C1 = {data['correlator_L_independence']['delta_C1']:.4f}")

print('
OS Axioms:')
for ax, status in data['os_axioms'].items():
    print(f'  {ax}: {status}')
```
